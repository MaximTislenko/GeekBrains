###В работу взята выборка по приемам врача-педиатра Тисленко М.В.Данная выписка содержит 1545 записей(строк), за период с 03.04.2020 по 18.09.2022. Для дальнейшего анализа производится переименовывание столбцов, и начинается работа с данными.

```python
data = data_first.copy()
data.rename(columns= {'Исполнитель':'physician',
                      'Пациент':'id_patient',
                      'ДатаРождения':'date_of_birth',
                      'Пол':'Sex',
                      'Номенклатура':'service',
                      'ДатаВыполнения':'visit_date',
                      'МКБ10Код':'ICD_10',
                      'ТипДиагноза':'diagnosis_type'},
             inplace= True)
```

###Переформатируется в формат дата-время с созданием нового столбцов: Дата рождения, Дата выполнения, для подсчета возраста в день обращения. 

```python
data['birthday'] = pd.to_datetime(data['date_of_birth'], errors= 'coerce', dayfirst= True)
data['visit_date_time'] = pd.to_datetime(data['visit_date'], errors= 'coerce', dayfirst= True)
data['age'] = (data['visit_date_time'] - data['birthday']) // np.timedelta64 (12, 'M')
```

Добавляется новый столбец для присваивания номера каждому пациенту.

```python
c = data['id_patient'].unique()
index_x = (1 + i for i in range( len(c)))
index_x = pd.DataFrame(index_x)
index_x.columns = ['id']
pation = pd.DataFrame(c)
pation.columns = ['id_patient']
pation = pd.concat([pation,index_x], sort= False, axis=1)
data = data.join(pation.set_index('id_patient'), on= 'id_patient')
```
