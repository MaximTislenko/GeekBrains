```python
import pandas as pd
import numpy as np
```

```python
data_first = pd.read_excel('/content/drive/MyDrive/Наставник/data.xlsx')
data_first
```

```python
data = data_first.copy()
data
```

```python
data.rename(columns= {'Исполнитель':'physician',
                      'Пациент':'patient',
                      'ДатаРождения':'date_of_birth',
                      'Пол':'Sex',
                      'Номенклатура':'service',
                      'ДатаВыполнения':'visit_date',
                      'МКБ10Код':'ICD_10',
                      'ТипДиагноза':'diagnosis_type'},
             inplace= True)
data
```

```python
data['birthday'] = pd.to_datetime(data['date_of_birth'], errors= 'coerce', dayfirst= True)
data['visit_date_time'] = pd.to_datetime(data['visit_date'], errors= 'coerce', dayfirst= True)
# data['age'] = (data['visit_date_time'] - data['birthday']) // np.timedelta64 (1, 'Y')
data['age'] = (data['visit_date_time'] - data['birthday']) // np.timedelta64 (12, 'M')
data
```
